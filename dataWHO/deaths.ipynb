#### Let's classify the risk factors in 4 categories to simplify further analysis

```python
risk_pollution = ['outdoor_air_pollution','household_air_pollution_from_solid_fuels','air_pollution']

risk_alimentary = ['diet_high_in_sodium','diet_low_in_whole_grains','diet_low_in_fruits','diet_low_in_nuts_and_seeds','diet_low_in_vegetables',
                   'vitamin_a_deficiency','iron_deficiency']

risk_hygene = ['unsafe_water_source','unsafe_sanitation','no_access_to_handwashing_facility']

risk_drug = ['alcohol_use','secondhand_smoke','smoking','drug_use']

risk_health = ['high_systolic_blood_pressure','high_fasting_plasma_glucose','high_body-mass_index','low_bone_mineral_density']

risk_early = ['low_birth_weight','child_wasting','child_stunting','discontinued_breastfeeding', 'non-exclusive_breastfeeding']

risk_ethologic = ['unsafe_sex','low_physical_activity']
```

```python
entity_test = 'Chile'
data[data['entity']== entity_test]
```

#### Let's look at the evolution of deaths related to pollution risks in Chile

```python
for risk in risk_pollution:
    data[data['entity']==entity_test].plot.line(
        x= 'year' , y= risk ,figsize=(12,3),lw=1.5, color = 'red',marker= 'o', title= 'Death associated to {risk} in {region}'.format(
        risk= risk , region= entity_test))
```

#### Observations:
* Household pollution has steadily become less life threatening. 
* Deaths releted to air and outdoor pollution have increased in the last 11 years (2008 - 2019


#### This results concerning!!!

```python
def intersection(list1, list2):
    list3 = [value for value in list1 if value in list2]
    return list3
```

#### IDEA #2: Study correlation between the time series

```python
risk_correlation = data.drop('year',axis=1).corr()
```

```python
plt.figure(figsize= (20,20))
mask = np.triu(np.ones_like(risk_correlation, dtype=bool))
heatmap = sns.heatmap(risk_correlation, vmin = -1, vmax= 1,mask= mask, annot = True, cmap = 'RdYlBu') #OBS: Decidir color
heatmap.set_title('Correlation Heatmap')
```

##### Maybe, if we look at the groups that were made earlier to look for 'blocks' of variables

```python
for lista in risk_list:
    risk_correlation = data.drop('year',axis=1)[lista].corr()
    mask = np.triu(np.ones_like(risk_correlation, dtype=bool))
    plt.figure(figsize= (12,12))
    heatmap = sns.heatmap(risk_correlation, vmin = -1, vmax= 1,mask=mask, annot = True, cmap = 'RdYlBu')
    heatmap.set_title('Correlation Heatmap of {risk_type} variables'.format(risk_type= lista))
```
